We are writing the following code to remove the outliers and focus on the main prices:

```python
lower_percentile = df['PRICE'].quantile(0.01)
upper_percentile = df['PRICE'].quantile(0.95)

# Filter the DataFrame based on the specified range
df = df[(df['PRICE'] > lower_percentile) & (df['PRICE'] < upper_percentile)]
```

***We can see that the dataset id pretty good, there are no null values, it gives all the info about the houses it is properly labelled and the values are consistent too, so we can start the analysis of the data***

# Scatterplot comparing the prices by sublocality:

```python
sublocalities = sublocalComparison.index
prices = sublocalComparison.values

plt.figure(figsize=(10, 5))
plt.scatter(sublocalities, prices, c='maroon')

plt.xlabel('Sublocality')
plt.ylabel('Total Price')
plt.title('Total Price by Sublocality')

plt.xticks(rotation=90)

# Show the plot
plt.show()
```

# Heatmap showing the correaltion between the price and size of the property:

```python
correlation = ['PRICE','PROPERTYSQFT']
correlationDf = df[correlation]
correlation_matrix = correlationDf.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correalation between the price and the size of the property")
plt.show()
```
